```python
#a.	Mean and std deviation of the income
print("Mean of income:", oj.INCOME.mean())
print("STD of income:", oj.INCOME.std())
```

```python
#b.	For income greater than or equal to 10.5, find the mean income

oj.INCOME[oj.INCOME >= 10.5].mean()
```

```python
#c.	For each brand having price >=2.5 find the mean, median, sd of the log of income
mean=oj[oj.price >=2.5] .groupby("brand").INCOME.mean()
print("Mean of Income:\n",mean)
print("\n\n")

median=oj[oj.price >=2.5] .groupby("brand").INCOME.median()
print("Median of Income:\n",median)

print("\n\n")

std=oj[oj.price >=2.5] .groupby("brand").LogInc.std()
print("Std of Log of Income:\n",std)
```
