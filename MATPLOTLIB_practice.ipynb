This document presents graphical analysis on the Crime Rate Dataset in India to FIND THE SAFEST STATE FOR FOREIGNERS.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
%matplotlib inline
import seaborn as sb
```

```python
df = pd.read_csv("C:\\Users\\Administrator\\Desktop\\Python Assignments\\DataSets\\clean.csv")
df['totalcases'] = df['TOTAL_Dacoity']+df['TOTAL_Robbery']+df['TOTAL_Burglary']+df['TOTAL_Theft']
df1 = df.groupby("STATE").agg({'totalcases': ['sum']}).reset_index()
a=pd.DataFrame(df1['STATE'])
b=pd.DataFrame(a.join(df1.totalcases))
b.rename(columns = {'STATE':'STATE','sum':'SUM'})
b.columns= ['STATE','SUM']
```

```python
b.sort_values(by = 'SUM').head(10).plot.bar(x='STATE',y='SUM',title = "SAFEST STATE FOR FOREIGNERS BASED ON TOTAL CASES IN INDIA")
```
