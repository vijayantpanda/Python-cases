```python
#How many unique cities are the orders being delivered to
store = pd.DataFrame(store)
#list(store)
a = store.City.unique()
b = len(a)
b
```

```python
#What is the total quantity sold in the East Region?

a = store[store.Region == 'East'].Quantity.count()
a

```

```python
#Find the sum of the quantity sold in the East Region
#a = store.query("(Region == 'East')")
#a.Quantity.sum()

a = store[store.Region == 'East'].Quantity.sum()
a
```

```python
#In the south region sort the sales in decreasing order
b = store[store.Region == 'South'].sort_values(by = 'Sales',ascending = [False])
b
```

```python
#Find the mean of quantity for every region
#mean = store[store.Quantity].groupby('Region').mean()

group = store.groupby('Region')
group['Quantity'].mean()
```

```python
#Find the mean of sales for every category
group = store.groupby('Category')
group['Sales'].mean()
```

```python
#Find the max, min, sum of sales and profit for every category

store.groupby('Category').Sales.agg(['max','min','sum'])

```

```python
# Find sum of sales and max profit for every segment
#store.groupby('Segment').Sales.agg(['sum'])
#print(store.Sales.sum())
#print(store.Profit.max())

store.groupby('Segment').agg({'Sales':np.sum,'Profit':np.max})
store.groupby('Segement'.agg({'Sales':np.sum,'Profit':np.max}))
```

```python
#For every segment find the mean of the discount
store.groupby('Segment').Discount.mean()
```

```python
# For every segment find the most profitable customers

a = store.groupby('Segment').Profit.max() 
b = store.Profit
store[["Customer Name","Segment","Profit"]][np.in1d(b,a)]

```

```python
# What are the top 5  sub categories that give maximum profit?
store.groupby('Sub-Category').Profit.max().nlargest(5)
```

```python
# What is the Total Sales, Quantity, Discount, Profit across Total US.
store[store.Country == 'United States'][["Sales","Quantity","Discount","Profit"]].agg(['sum'])
```

```python
#How many times has it taken more than 5 days from placing an order to shipping store.

store['Order Date'] = pd.to_datetime(store['Order Date'])
store['Ship Date'] = pd.to_datetime(store['Ship Date'])
sum(store['Ship Date'] - store['Order Date'] > '5 days')
```

```python
# Find the total number of orders in every category which has been shipped with a duration > 5 days

store[(store['Ship Date'] - store['Order Date'] > '5 days')].groupby('Category')[["Order ID"]].count()
```

```python
#What’s the percentage of items which has been shipped within 5 days

sum((store['Ship Date'] - store['Order Date']) < ' 5 days')* 100/len(store['Order Date'])
```

```python
#What’s the percentage of items which has been shipped after 5 days
sum((store['Ship Date'] - store['Order Date']) > ' 5 days')* 100/len(store['Order Date'])
```
